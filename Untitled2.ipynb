```python
import numpy as np
import matplotlib.pyplot as plt

F0 = 50        # frecuencia de la señal (Hz)
Fs = 5*F0      # frecuencia de muestreo (>= 2*F0)

t = np.linspace(0, 0.04, 1000)   # tiempo continuo
x = np.sin(2*np.pi*F0*t)         # señal continua

# tiempos discretos
n = np.arange(0, 0.04, 1/Fs)
x_d = np.sin(2*np.pi*F0*n)

# graficar
plt.figure()
plt.plot(t, x, label="Señal continua")
plt.stem(n, x_d, linefmt='r-', markerfmt='ro', basefmt=" ", label="Muestras")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.show()
```
